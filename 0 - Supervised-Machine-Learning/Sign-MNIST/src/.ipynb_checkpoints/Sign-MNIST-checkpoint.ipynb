In this notebook, we will build a model on the sign MNIST dataset.

We load the **digits dataset** from sklearn.

```python
# Imports
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import seaborn as sns
import string
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D,Dropout
from sklearn.metrics import classification_report, confusion_matrix
```

```python
train_data = pd.read_csv('../data/sign_mnist_train.csv')
test_data = pd.read_csv('../data/sign_mnist_test.csv')
```

As usual, let's have a quick look at the data.

```python
X_train = np.array(train_data.drop('label', axis=1))
y_train = np.array(train_data['label'])
X_test = np.array(test_data.drop('label', axis=1))
y_test = np.array(test_data['label'])
```

```python
print('Shape of X_train : {} , Shape of X_test : {}' .format(X_train.shape,X_test.shape))
print('Shape of y_train : {} , Shape of y_test : {}' .format(y_train.shape,y_test.shape))
```

```python
set(y_train)
```

```python
#dictionnaire key = label and value = letters
num_letters = {0 : 'A' , 1 : 'B', 2 : 'C' , 3 : 'D', 4: 'E' , 5 : 'F',
                     6 : 'G' , 7 : 'H' , 8 : 'I' , 10 : 'K' , 
                     11 : 'L' , 12 :'M' , 13 : 'N' , 14 : 'O' , 15 : 'P',
                     16 : 'Q' , 17 : 'R' , 18 : 'S' , 19 :'T' , 20 : 'U',
                     21 : 'V' , 22 :'W' , 23 : 'X', 24 : 'Y'}

plt.figure(figsize = (10,10))

for i in range(9):
    
    ind = random.randint(0, X_train.shape[0])
    
    plt.subplot(3,3,i+1)
    plt.imshow(X_train[ind].reshape(28,28))
    
    plt.title(num_letters[y_train[ind]], fontdict={'color' : 'Black', 'fontsize': 20})
    
plt.show()
```

```python
# Is the train dataset unbalanced  ?
sns.displot(y_train)
```

```python
# we are going to use data generator to train the CNN
X_train = X_train/255
X_test = X_test/255


# transforming the labels (dummyfying)

y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
# CNN building
model = Sequential()

model.add(Conv2D(filters = 256,kernel_size = (5,5),padding = 'same',activation = 'relu',input_shape = (28,28,1)))
model.add(MaxPooling2D())
model.add(Conv2D(filters = 128,kernel_size = (5,5),padding = 'same',activation = 'relu'))
model.add(MaxPooling2D())
model.add(Conv2D(filters = 64,kernel_size = (5,5),padding = 'same',activation = 'relu'))
model.add(MaxPooling2D())
model.add(Conv2D(filters = 32,kernel_size = (5,5),padding = 'same',activation = 'relu'))


model.add(Flatten())

model.add(Dense(units = 256,activation = 'relu'))
model.add(Dense(units = 128,activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(25,activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',metrics = ['accuracy'],loss = 'categorical_crossentropy')
```

```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

```python
model.fit(x = X_train, y = y_train, batch_size = 128, validation_data = (X_test,y_test), epochs = 5)
```

```python
y_pred = model.predict(X_test)
```

### Confusion matrix

```python
y_pred_num_classes = np.argmax(y_pred, axis=1)
y_test_num_classes = np.argmax(y_test, axis=1)
```

```python
y_pred_letters = [num_letters[y_pred_num_classes[i]] for i in range(0, len(y_pred_num_classes))]
```

```python
y_test_letters = [num_letters[y_test_num_classes[i]] for i in range(0, len(y_test_num_classes))]
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(confusion_matrix(y_test_letters, y_pred_letters), annot= True,cbar = False , cmap = 'Blues',annot_kws={'size' : 8})
plt.xticks(ticks=np.arange(0,24) + 0.5, labels=[i for i in list(string.ascii_uppercase) if i not in ['J','Z']])
plt.yticks(ticks=np.arange(0,24) + 0.5, labels=[i for i in list(string.ascii_uppercase) if i not in ['J','Z']])
plt.xlabel('Predicted',fontdict={'size': 20})
plt.ylabel('True' ,fontdict={'size': 20})
plt.title('Confusion Matrix',fontdict={'size': 20,'color':'Blue'})
plt.show()
```

### Let's see some random predictions

```python
plt.figure(figsize = (10,10))

for i in range(9):
    
    ind = random.randint(0, X_test.shape[0])
    
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[ind].reshape(28,28))
    
    plt.title(f'pred:{num_letters[np.argmax(y_pred, axis=1)[ind]]} / true:{num_letters[np.argmax(y_test, axis=1)[ind]]}', fontdict={'color' : 'Black', 'fontsize': 20})
    
plt.show()
```

### Let's now see some random wrong predictions

```python
errors_indices = [i for i in range(0, len(y_test)) if np.argmax(y_pred,axis=1)[i] != np.argmax(y_test,axis=1)[i]]
```

```python
plt.figure(figsize = (10,10))

for i in range(9):
    
    ind = random.choice(errors_indices)
    
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[ind].reshape(28,28))
    
    plt.title(f'pred:{num_letters[np.argmax(y_pred, axis=1)[ind]]} / true:{num_letters[np.argmax(y_test, axis=1)[ind]]}', fontdict={'color' : 'Black', 'fontsize': 20})
    
plt.show()
```
