### II.1. Encodings

Usually, when you want to make a face recognition, you already know who you want to recognize. This is our case here. So, instead of computing the encodings of all the people you want to recognize each time, you do it once at the beginning.

This is your first step: compute the encodings of each picture. 

Store the encodings of each member of the family into a dictionary of 4 members (with their names as keys), containing each a list of 4 encodings (one for each picture of the member).

```python
stark_names = ['arya', 'jon', 'robb', 'sansa']
```

```python
os.listdir('train/arya')
```

```python
# TODO: Compute the encodings of each member
stark_encodings = {}
train_pics = {}
for name in stark_names:
    train_pics[name] = [plt.imread(f'train/{name}/{file}') for file in os.listdir(f'train/{name}')]
```

```python
for name in stark_names:
    stark_encodings[name] = [face_recognition.face_encodings(pic)[0] for pic in train_pics[name]]
```

```python
stark_encodings
```

```python
def find_starks(img_path, threshold=0.55):
    img = plt.imread(img_path)
    
    encoding_img = face_recognition.face_encodings(img)
    location_img = face_recognition.face_locations(img)
    
    labels_lst = []
    for i in range(len(encoding_img)):
        for key, value in stark_encodings.items():
            if True in face_recognition.compare_faces(stark_encodings[key], encoding_img[i], tolerance=threshold):
                labels_lst.append([key, location_img[i]])
                break
    
    fig, ax = plt.subplots()
    plt.imshow(img)
    for label in labels_lst:
        ymin, xmin, ymax, xmax = label[1]
        ax.add_patch(matplotlib.patches.Rectangle((xmin, ymin), (xmax-xmin), (ymax-ymin), fill=False, edgecolor='red'))
        ax.text(label[1][1]-40, label[1][0]-10, label[0], color='red')
    plt.axis('off')
    plt.show()
    
    return labels_lst
```
