# Challenge - Boston House Prices

## 1. Data Preparation

```python
import pandas as pd
```

```python
# Import and load the dataset
from sklearn.datasets import load_boston

X, y = load_boston(return_X_y=True)
```

```python
# Have a look at the data
```

```python
X_df = pd.DataFrame(X, columns=load_boston().feature_names)
```

```python
X_df
```

```python
y
```

```python
# Let's split and rescale
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## 2. Modelling

Now we want to build a Neural Network to perform this regression task. We will build a Neural Networks with 5 hidden layers of 100 units each.

```python
# Create the Sequential object
import tensorflow as tf
```

```python
def multilayer_perceptron(input_dim):
    
    model = tf.keras.models.Sequential()
    
    model.add(tf.keras.layers.Dense(100, input_dim=input_dim, activation='sigmoid'))
    model.add(tf.keras.layers.Dense(100, activation='sigmoid'))
    model.add(tf.keras.layers.Dense(100, activation='sigmoid'))
    model.add(tf.keras.layers.Dense(100, activation='sigmoid'))
    model.add(tf.keras.layers.Dense(100, activation='sigmoid'))
    
    model.add(tf.keras.layers.Dense(1, activation='linear'))
    
    return model
```

```python
mlp = multilayer_perceptron(13)
```

```python
mlp.summary()
```

Now let's compile the model and then fit it

```python
# Compile the model with mean squared error (for regression)
mlp.compile(loss='mean_squared_error', metrics=[tf.keras.metrics.MeanAbsoluteError()])


# Now fit the model on 500 epoches with a batch size of 64
mlp.fit(X_train, y_train, epochs=500, batch_size=64, validation_data=(X_test, y_test))

# You can add the test/validation set into the fit: it will give insights on this dataset too
```

Finally we can compute the results of our model

```python
# Evaluation
loss, mae = mlp.evaluate(X_test, y_test, verbose=0)
print('RMSE is:', loss**(1/2))
print('MAE is:', mae)
```

## 3. Benchmarking

We can compare that result to a good old linear regression

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
rmse = mean_squared_error(y_test, y_pred)**(1/2)
mae = mean_absolute_error(y_test, y_pred)

print('RMSE is:', rmse**(1/2))
print('MAE is:', mae)
```

Of course to compare properly the two models, one would make hyperparameter optimization first, this is just for the example.
