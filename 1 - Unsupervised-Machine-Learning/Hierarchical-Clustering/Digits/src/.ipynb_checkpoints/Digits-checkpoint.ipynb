# Digits ✍️

We will work on the digits dataset and try to classify the data using clustering algorithms!

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

### Let's load the dataset, rescale it and use PCA(32) then tSNE(2) to obtain sharp results.

```python
# Load dataset (limit to 500 entries)
X,y = load_digits(return_X_y=True)[:500]
```

```python
# Rescaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_sc = scaler.fit_transform(X)
```

```python
# Apply dimension reduction
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
pca = PCA(n_components=32)
X_pca = pca.fit_transform(X_sc)
```

```python
tsne = TSNE(n_components=2)
X_final = tsne.fit_transform(X_pca)
```

```python
X_final
```

### Plot the data points

```python
# Plot data points
plt.figure()

plt.scatter(X_final[:,0], X_final[:,1], c=y)

plt.show()
```

### Let's use AgglomerativeClustering to cluster the data 

```python
# Generate dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
linked = linkage(X_final, method='single')
```

```python
plt.figure(figsize=(15,15))

dendrogram(linked)

plt.title('Dendrogram of digits dataset with PCA(32) TSNE(2)')

plt.show()
```

```python
# Apply sklearn AgglomerativeClustering algorithm on processed data
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=10)
ac.fit(X_final)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(121)
plt.scatter(X_final[:,0], X_final[:,1], c=y)

plt.subplot(122)
plt.scatter(X_final[:,0], X_final[:,1], c=ac.labels_)

plt.show()
```

### Evaluating the classification model

```python
ac.fit_predict(X_final)
```

```python
# Affect label to cluster
import statistics as stat
y_pred = [stat.mode(y[np.where(ac.labels_==ac.labels_[i])]) for i in range(X_final.shape[0])]
```

```python
# Compute score
```

```python
score = (y - y_pred == 0).mean()
```

```python
print(f'Score : {score:.2%}')
```
